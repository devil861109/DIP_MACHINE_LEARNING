## Hacer un diagrama de dispersión con listas
En este ejercicio, utilizaremos un conjunto de datos que contiene información sobre 227 países. Este conjunto de datos tiene mucha información interesante sobre cada país, como las tasas de natalidad, las tasas de mortalidad y su producto interno bruto (PIB). El PIB es el valor de todos los bienes y servicios producidos en un año, expresado en dólares por persona.

Hemos creado tres listas de datos de este conjunto de datos para que pueda comenzar. `gdp` es una lista que contiene el valor del PIB por país, expresado en dólares por persona. teléfonos es una lista del número de `phones` móviles por cada 1.000 personas en ese país. Finalmente, `percent_literate` es una lista que contiene el porcentaje de la población de cada país que sabe leer y escribir.


* Importe Matplotlib y Seaborn utilizando la convención de nomenclatura estándar.

```python
# Importar Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Esto solo funciona con colab
from google.colab import drive
#drive.mount('/content/drive/MyDrive/Colab Notebooks/Machine Learning/04 Introduccion Seaborn')
drive.mount("/content/drive")
path = "/content/drive/MyDrive/Colab Notebooks/Machine Learning/04 Introduccion Seaborn/countries-of-the-world.csv"
countries = pd.read_csv(path)
```

```python
# Standalone
# countries = pd.read_csv('datasets/countries-of-the-world.csv')
```

* Cree un diagrama de dispersión del PIB (gdp) frente al número de teléfonos por cada 1000 personas (phones).

```python
countries = countries.dropna()
phones = countries['Phones (per 1000)'].to_list()
gdp = countries['GDP ($ per capita)'].to_list()
percent_literate = countries['Literacy (%)'].to_list()
# Cree un diagrama de dispersión con el PIB en el eje x y el número de teléfonos en el eje y
sns.scatterplot(x=gdp, y=phones)
sns.despine()
```

* Cambie el diagrama de dispersión para que muestre el porcentaje de la población que sabe leer y escribir (percent_literate) en el eje y.

```python
# Cree un diagrama de dispersión con el PIB en el eje x y el número de teléfonos en el eje y
sns.scatterplot(x=gdp, y=percent_literate)
sns.despine()
```

¡Está bien! Si bien este gráfico no muestra una relación lineal entre el PIB y el porcentaje de alfabetización, los países con un PIB más bajo parecen tener más probabilidades de tener un porcentaje más bajo de población que sabe leer y escribir.

## Hacer un gráfico de conteo con una lista
En el último ejercicio, exploramos un conjunto de datos que contiene información sobre 227 países. Exploremos más estos datos: específicamente, ¿cuántos países hay en cada región del mundo?

Para hacer esto, necesitaremos usar un gráfico de conteo. Los gráficos de conteo incluyen una lista categórica y devuelven barras que representan el número de entradas de la lista por categoría. Puede crear uno aquí usando una lista de regiones para cada país, que es una variable llamada `region`.

* Importe Matplotlib y Seaborn utilizando las convenciones de nomenclatura estándar.
* Utilice Seaborn para crear un gráfico de recuento con una región en el eje y.
* Muestra la trama.

```python
# Importar Matplotlib y Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

region = countries['Region']

# Crear gráfico de recuento con región en el eje y
sns.countplot(y=region)

# Mostrar trama
plt.show()
```

¡Gran trabajo! El África subsahariana contiene la mayor cantidad de países en esta lista. Volveremos a examinar los gráficos de conteo más adelante en el curso.

## Diagramas de tono y dispersión
Anteriormente, aprendimos cómo el tono nos permite crear subgrupos fácilmente dentro de los gráficos de Seaborn. Probémoslo explorando datos de estudiantes de secundaria. Tenemos mucha información sobre cada estudiante como su edad, dónde vive, sus hábitos de estudio y sus actividades extracurriculares.

Por ahora, veremos la relación entre el número de ausencias que tienen en la escuela y su calificación final en el curso, segmentado por el lugar donde vive el estudiante (zona rural vs. urbana).

* Cree un diagrama de dispersión con "ausencias" en el eje x y calificación final ("G3") en el eje y usando el DataFrame Student_data. Colorea los puntos de la trama según la "ubicación" (urbana o rural).

```python
path3 = "/content/drive/MyDrive/Colab Notebooks/Machine Learning/04 Introduccion Seaborn/student-alcohol-consumption.csv"
student_data = pd.read_csv(path3)

# Cree un diagrama de dispersión de ausencias versus calificación final
sns.scatterplot(x="absences", y="G3",
                data=student_data,
                hue="location")

# Mostrar trama
plt.show()
```

* Haz que "Rural" aparezca antes de "Urbano" en la leyenda de la trama.

```python
# Cambiar el orden de las leyendas en el diagrama de dispersión.
sns.scatterplot(x="absences", y="G3",
                data=student_data,
                hue="location",
                hue_order=["Rural", "Urban"])

# Mostrar trama
plt.show()
```

¡Bien hecho! Parece que los estudiantes con mayores ausencias tienden a tener calificaciones más bajas tanto en áreas rurales como urbanas.

## Gráficos de tono y recuento
Sigamos explorando nuestro conjunto de datos de estudiantes de secundaria observando una nueva variable. La columna "escuela" indica las iniciales de la escuela a la que asistió el estudiante: "GP" o "MS".

En el último ejercicio, creamos un diagrama de dispersión donde los puntos del diagrama se colorearon según si el estudiante vivía en un área urbana o rural. ¿Cuántos estudiantes viven en áreas urbanas versus rurales? ¿Esto varía según la escuela a la que asiste el estudiante? Hagamos un gráfico de conteo con subgrupos para averiguarlo.

* Complete el diccionario paleta_colores para asignar el valor de ubicación "Rural" al color "verde" y el valor de ubicación "Urbana" al color "azul".
* Cree un gráfico de recuento con "escuela" en el eje x utilizando el marco de datos `student_data`.
  + Agregue subgrupos al gráfico usando la variable "ubicación" y use el diccionario `palette_colors` para hacer que los subgrupos de ubicación sean verdes y azules.

```python
# Crear un diccionario que asigne valores de subgrupo a colores
palette_colors = {"Rural": "green", "Urban": "blue"}

# Cree un gráfico de conteo de la escuela con subgrupos de ubicación
sns.countplot(x="school", data=student_data,
              hue="location",
              palette=palette_colors)

# Mostrar gráfico
plt.show()
```

Impresionante. Los estudiantes de GP tienden a provenir de una ubicación urbana, pero los estudiantes de MS están divididos de manera más equitativa.
